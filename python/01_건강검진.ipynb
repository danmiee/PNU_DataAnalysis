```python
df['성별코드']
```

```python
df2 = df[["성별코드","허리둘레","수축기 혈압","이완기 혈압","식전혈당(공복혈당)","트리글리세라이드","HDL 콜레스테롤"]]
```

```python
df2.columns = ["성별코드","허리둘레","수축기혈압","이완기혈압","공복혈당","트리글리세라이드","HDL콜레스테롤"]
```

```python
lt = list(df2.columns)
print('lt1: ',lt)
lt = [c.replace(' ','') for c in lt]
print('lt2: ',lt)
lt[4] = '공복혈당'
print('lt3: ',lt)
```

```python

df2.columns = lt
df2.info()
```

```python
df2 = df2.dropna(axis=0)
df2.info()
```

```python
import numpy as np
```

# 높은 혈압(130/85mmHg 이상)
df2$높은혈압 <- ((df2$수축기혈압 >= 130) | (df2$이완기혈압 >= 85))

```python
df2['높은혈압'] = np.where((df2['수축기혈압']>=130) | (df2['이완기혈압']>=85),1,0)
```

```python
df2.head()
```


# 높은 혈당(공복 혈당 100mg/dL 이상)
df2$높은혈당 <- df2$공복혈당 >= 100

```python
df2['높은혈당'] = np.where(df2['공복혈당']>=100,1,0)
```

```python
df2.head()
```

# 높은 중성지방(트리글리세라이드 150mg/dL 이상)
df2$높은중성지방 <- df2$트리글리세라이드 >= 150

```python
df2['높은중성지방'] = np.where(df2['트리글리세라이드']>=150,1,0)
```

```python
df2.head()
```

# 낮은 HDL 콜레스테롤 수치(남성은 40mg/dL 미만, 여성은 50mg/dL 미만)
df2$낮은콜레스테롤 <- (((df2$성별코드 == 1) & (df2$HDL콜레스테롤 < 40)) | ((df2$성별코드 == 2) & (df2$HDL콜레스테롤 <50)))

```python
df2['낮은콜레스테롤'] = np.where(
  ((df2['성별코드']==1) & (df2['HDL콜레스테롤']<40)) | 
  ((df2['성별코드']==2) & (df2['HDL콜레스테롤']<50)),1,0)
```

```python
df2.head()
```

# 복부 비만(남성 90cm 이상, 여성 85cm 이상)
df2$복부비만 <- (((df2$성별코드 == 1) & (df2$허리둘레 >= 90)) | ((df2$성별코드 == 2) & (df2$허리둘레 >= 85)))

```python
df2['복부비만'] = np.where(
  ((df2['성별코드']==1) & (df2['허리둘레']>=90)) | 
  ((df2['성별코드']==2) & (df2['허리둘레']>=85)),1,0)
```

```python
df2.head()
```

# 대사증후군 수
df2$대사증후군 <- (df2$높은혈압 + df2$높은혈당 + df2$높은중성지방 + df2$낮은콜레스테롤 + df2$복부비만)

```python
df2['대사증후군'] = df2['높은혈압']+df2['높은혈당']+df2['높은중성지방']+df2['낮은콜레스테롤']+df2['복부비만']
```

```python
df2.head()
```

df2$성별 <- ifelse(df2$성별코드 == 1, "남", "여")

```python
df2['성별'] = np.where(df2['성별코드']==1,"남","여")
```

```python
df2.head()
```

# 판별
# 0 : 정상, 1~2 : 주의군, 3~5 : 위험군
df2$판별 <- ifelse(df2$대사증후군 == 0, "정상", ifelse(df2$대사증후군 <= 2, "주의군", "위험군"))

table(df2$판별, df2$성별)

```python
df2['판별'] = np.where(df2['대사증후군']==0,"정상",
                     np.where(df2['대사증후군']<=2,"주의군","위험군"))
```

```python
df2.head()
```

```python
df3 = df2.groupby(['성별', '판별']).count()['성별코드']
```
